```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('year_rev.csv')
df1.head()
```

```python
df2 = pd.read_csv('year_mean_length.csv')
df2.head()
```

```python
df = df1.merge(df2, on ='year')
df.columns = ['year', 'review', 'text1']
df['totalText'] = df.review*df.text1
df.head()
```

```python
sns.regplot(df.year, df.totalText)
plt.title ("Total words vs year")
plt.xlabel("Year")
plt.ylabel("Words")
```

```python
sns.set_style("dark")
p = sns.regplot(df.year, df.totalText, order = 2)
plt.title ("Total texts vs year")
plt.xlabel("year")
plt.ylabel("average number of words")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.io import gbq
%matplotlib inline
```

```python
query1 = "SELECT * FROM amazon_book_small.rating \
WHERE unixReviewTime >= '1996-01-01 00:00:00 UTC' AND unixReviewTime < '2013-01-01 00:00:00 UTC'"
```

```python
project_id = 'dotted-chiller-156222'
```

```python
da = gbq.read_gbq(query1, project_id=project_id)
```

```python
da.head(1)
```

```python
query2 = "SELECT * FROM amazon_book_small.rating \
WHERE unixReviewTime >= '2013-01-02 00:00:00 UTC' AND unixReviewTime < '2014-01-01 00:00:00 UTC'"
```

```python
db = gbq.read_gbq(query2, project_id=project_id)
```

```python
db.head(1)
```

```python
query3 = "SELECT * FROM amazon_book_small.rating \
WHERE unixReviewTime >= '2014-01-02 00:00:00 UTC' AND unixReviewTime < '2015-01-01 00:00:00 UTC'"
```

```python
dc = gbq.read_gbq(query3, project_id=project_id)
```

```python
dc.head(1)
```

```python
df = pd.concat([da, db, dc])
df.head(1)
```

```python
df.columns = ['unixReviewTime', 'totalVote', 'wordLength', 'rating']
```

```python
df.head(1)
```

```python
sns.set_style("darkgrid")
year = df.groupby([df.unixReviewTime.dt.year, df.rating]).size()
year = year.unstack().fillna(0)
year.plot(marker = 'o', markersize = 6)
plt.xlabel('Year')
plt.ylabel('Number of reviews')
plt.show()
```

```python
a = df.groupby([df.unixReviewTime.dt.year, 'rating'])
b = a['wordLength'].mean()
b = b.unstack().fillna(0)
b.plot(marker = 'o', markersize = 6)
plt.xlabel('Year')
plt.ylabel('Average text length (words)')
plt.show()
```
